#### Select top 11 featurs for your classification [2]

```python
# We will do this dynamically
# first, we reset the index of our score matrix so we can access those feature names
scores = corr_matrix_labels.reset_index()

# next, order the rows based on their absolute values, in descending order based on those absolute score values
scores_ordered = scores.reindex(scores["class"].abs().sort_values(ascending=False).index)

# now to select the top 11 features based on the correlation scores
features_top_11 = all_data[scores_ordered["index"][:11]].drop(columns=["index"])
print(features_top_11.columns)
```

#### Using a virtualization method that you deem appropriate, compare local and imported beverage by their  properties [11]

### Building a models
#### Separate x and y data [2]

```python
features = prepped_data.drop(columns=["class"])
labels = prepped_data["class"]
```

#### Get the training and testing sets. Use 25% of the data for testing [5]

```python
features = StandardScaler().fit_transform(features)
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.25)
```

#### build a pipeline with standard scaler followed by logistic regression and fit the model [10]

```python
pipe_logger = Pipeline([('scaler', StandardScaler()), ('logger', LogisticRegression())]).fit(x_train, y_train)
```

 #### build a pipeline with standard scaler followed by Random Forest and fit the model [5]

```python
pipe_rando = Pipeline([('scaler', StandardScaler()), ('rando', RandomForestClassifier(n_estimators=100))]).fit(x_train, y_train)
```

#### build a pipeline with standard scaler followed by  _Another Model of Your Choice_  and fit the model [5]

```python
pipe_perco = Pipeline([("scaler", StandardScaler()), ('perceptron', Perceptron(max_iter=2000))]).fit(x_train, y_train)
```

#### Evaluate each model using:
1. A confusion matrix to see how the model's predictions align with the actual class labels.**[2 x 3]**
2. Accuracy to tell how many the model got right. **[1 x 3]**

```python
pipelines = {"Logistic Regression": pipe_logger,"Random Forest": pipe_rando,"Perceptron": pipe_perco}

for name, pipe in pipelines.items():
    y_pred = pipe.predict(x_test)
    print("\n=========================================================\n")
    print(f"Score Sheet for {name} model:\n")

    cm = confusion_matrix(y_test, y_pred, labels=pipe.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Local", "Import"])
    disp.plot()
    plt.show()

    
    print(classification_report(y_test, y_pred))
    print(f"Score Model on Data: {round(pipe.score(x_test, y_test), 2)}")
    print(f"ROC Area Under Curve: {round(roc_auc_score(y_test, y_pred), 2)}\n")

    fpr, tpr, thresholds = roc_curve(y_pred, y_test)
    plt.plot(fpr, tpr)
    plt.plot(fpr, fpr, linestyle = '--', color = 'k')
    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.title('ROC curve')
    plt.show()
   
```
